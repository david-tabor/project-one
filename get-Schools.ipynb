```python
# Import dependencies
import requests
import pandas as pd
from pprint import pprint
from sodapy import Socrata
```

```python
# Retrieve data using Socrata function provided by sodapy (API developers)
client = Socrata("data.cityofchicago.org", None)
schools = client.get("d2h8-2upd", limit=500000)

# Convert  results to pandas DataFrame
schools_df = pd.DataFrame.from_records(schools)

schools_df.head()
```

```python
# Extract useful rows into 'df' which will be saved to csv after munging
revDF = schools_df[['short_name', 'school_id', 'governance', 'address', 'commarea', 'zip', 'lat', 'long', 
                    'grade_cat', 'grades']]




revDFtrimmed = revDF.loc[(revDF['grades'] == 'K, 1, 2, 3, 4, 5, 6, 7, 8') | (revDF['grades'] == '9, 10, 11, 12')]
revDFtrimmed =  revDFtrimmed.reset_index()
revDFtrimmed
```

```python
#Save output to csv
revDFtrimmed.to_csv('schools.csv', index=False)
```
